```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("results_random.csv")
df = df.replace(
    {
        "morganfp": "MorganFP",
        "schnet_emb": "SchNet",
        "dimenet_emb": "DimeNet",
        "dimenetpp_emb": "DimeNet++",
        "molformer": "MolFormer",
        "lasso": "Lasso",
        "ridge": "Ridge",
        "xgb": "XGBoost",
        "lgbm": "LightGBM",
        "rf": "RF",
    }
)
```

```python
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].set_ylim(0, 5)
axes[0].set_title("Train RMSE")
sns.barplot(
    x="model_name",
    y="rmse_train",
    hue="desc_name",
    data=df,
    palette="muted",
    ax=axes[0],
)
axes[1].set_ylim(0, 5)
axes[1].set_title("Test RMSE")
sns.barplot(
    x="model_name", y="rmse_test", hue="desc_name", data=df, palette="muted", ax=axes[1]
)
```
