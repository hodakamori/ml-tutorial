```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_csv("../data/delaney-mordred.csv")
train, test = train_test_split(df, random_state=42, test_size=0.2)
```

```python
train
```

```python
from pycaret.regression import setup
exp = setup(train, target='logS', ignore_features=["name", "smiles"])
```

```python
exp.compare_models(exclude=["mlp", "gbr", "xgboost", "lightgbm"])
```

```python
base_model = exp.create_model('et')
tuned_model = exp.tune_model(base_model, n_iter=5, optimize='mae')
```

```python
exp.evaluate_model(tuned_model)
```

```python
final_model = exp.finalize_model(tuned_model)
test_pred = exp.predict_model(final_model, data=test)
```
