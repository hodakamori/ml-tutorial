```python
import pandas as pd

df = pd.read_csv("../data/delaney-mordred.csv")
df
```

```python
X = df.loc[:, df.columns.str.startswith("mordred_")]
X
```

```python
y = df["logS"]
y
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_std = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_std, y)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
import numpy as np

kf = KFold(n_splits=5, random_state=42, shuffle=True)
params = np.linspace(0, 1, 11)
scores = []
for alpha in params:
    mae = []
    for i, (train_index, val_index) in enumerate(kf.split(X_train)):
        X_tr, X_val = X_train[train_index], X_train[val_index]
        y_tr, y_val = y_train.iloc[train_index], y_train.iloc[val_index]
        model = Lasso(alpha=alpha)
        model.fit(X_tr, y_tr)
        y_val_pred = model.predict(X_val)
        mae_val = mean_absolute_error(y_val, y_val_pred)
        print(f"alpha: {alpha}, fold: {i},  MAE= {mae_val}")
        mae.append(mae_val)
    scores.append(sum(mae) / len(mae))
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(params, scores, marker="o", ms=5)
ax.set_xlabel(r"$\alpha$")
ax.set_ylabel("MAE")
```

```python
result = pd.DataFrame({"alpha": params, "score": scores})
result
```

```python
best_alpha = result.loc[result["score"].argmin(), "alpha"]
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

model = Lasso(alpha=best_alpha)
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
print(f"R2(train): {r2_train}, R2(test): {r2_test}")
print(f"MAE(train): {mae_train}, MAE(test): {mae_test}")
```
