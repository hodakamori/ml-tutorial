```python
import pandas as pd
df = pd.read_csv("../data/delaney-mordred.csv")
df
```

```python
df = df.dropna()
df
```

```python
X = df.loc[:, df.columns.str.startswith("mordred_")]
X
```

```python
y = df["logS"]
y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_std = sc.fit_transform(X)
```

```python
X_std_df = pd.DataFrame(X_std)
X_std_df.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std, y)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
print(f"R2(train): {r2_train}, R2(test): {r2_test}")
print(f"MAE(train): {mae_train}, MAE(test): {mae_test}")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(y_train, y_train_pred, s=5, alpha=0.5, label="train")
ax.scatter(y_test, y_test_pred, s=5, alpha=0.5, label="test")
ax.set_xlabel("logS(true)")
ax.set_ylabel("logS(pred)")
ax.set_xlim(y.min(), y.max())
ax.set_ylim(y.min(), y.max())
ax.set_aspect("equal")
```

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha=0.2)
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
print(f"R2(train): {r2_train}, R2(test): {r2_test}")
print(f"MAE(train): {mae_train}, MAE(test): {mae_test}")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(y_train, y_train_pred, s=5, alpha=0.5, label=f"train, R2={round(r2_train,2)}, MAE={round(mae_train,2)}")
ax.scatter(y_test, y_test_pred, s=5, alpha=0.5, label=f"test, R2={round(r2_test,2)}, MAE={round(mae_test,2)}")
ax.plot([-10, 5], [-10, 5], linewidth=1, color="tab:gray")
ax.set_xlabel("logS(true)")
ax.set_ylabel("logS(pred)")
ax.set_xlim(y.min(), y.max())
ax.set_ylim(y.min(), y.max())
ax.set_aspect("equal")
ax.set_xlim(-10, 5)
ax.set_ylim(-10, 5)
ax.legend()
ax.grid()
```

```python
coeff_df = pd.DataFrame({"coeff":model.coef_})
coeff_df["coeff_abs"] = abs(coeff_df["coeff"])
coeff_df = coeff_df.sort_values("coeff_abs", ascending=False)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
top20_coeff = coeff_df.iloc[:20]
ax.barh([f"mordred_{i}" for i in top20_coeff.index], top20_coeff["coeff"])
ax.axvline(x=0, color="tab:gray", linewidth=1)
ax.set_xlim(-1, 1)
ax.invert_yaxis()
ax.set_xlabel("Coefficient")
```
