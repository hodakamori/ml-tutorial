```python
import pandas as pd
df = pd.read_csv("../data/delaney-mordred.csv")
```

```python
df
```

```python
base_feature = df.loc[:, [f"mordred_{i}" for i in range(10)]]
```

```python
import featuretools as ft
es = ft.EntitySet(id='esol')
es = es.add_dataframe(
    dataframe_name="esol",
    dataframe=base_feature,
)
```

```python
feature_matrix, feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name="esol",
    trans_primitives=['add_numeric', 'subtract_numeric', "multiply_numeric", "divide_numeric"],
    agg_primitives=[],
    max_depth=1
)
feature_matrix
```
