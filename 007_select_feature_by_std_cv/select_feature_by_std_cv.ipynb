```python
import pandas as pd

df = pd.read_csv("../data/delaney-mordred.csv")
df = df.dropna()
df
```

```python
y = df["logS"]
y
```

```python
X = df.loc[:, df.columns.str.startswith("mordred_")]
X
```

```python
std_thresh = 10
X_slice_std = X.loc[:, X.std() > std_thresh]
X_slice_std
```

```python
import numpy as np
import matplotlib.pyplot as plt

std_thresh_list = np.linspace(0, 20, 21)
num_features = []
for std_thresh in std_thresh_list:
    X_slice_std = X.loc[:, X.std() > std_thresh]
    num_features.append(X_slice_std.shape[1])

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(std_thresh_list, num_features, marker="o", ms=4)
ax.set_xlabel("STD threshold")
ax.set_ylabel("Number of features")
ax.set_ylim(0, X.shape[1] + 10)
ax.grid(True)
```

```python
cv_thresh_list = np.linspace(0, 10, 21)
num_features = []
for cv_thresh in cv_thresh_list:
    X_slice_cv = X.loc[:, X.std() / X.mean() > cv_thresh]
    num_features.append(X_slice_cv.shape[1])

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(cv_thresh_list, num_features, marker="o", ms=4)
ax.set_xlabel("CV threshold")
ax.set_ylabel("Number of features")
ax.set_ylim(0, X.shape[1] + 10)
ax.grid(True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error


def train_and_eval(X, y):

    X_train, X_test, y_train, y_test = train_test_split(X, y)
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    mae_train = mean_absolute_error(y_train, y_train_pred)
    mae_test = mean_absolute_error(y_test, y_test_pred)

    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)
    print(f"R2(train): {r2_train:.2f}, R2(test): {r2_test:.2f}")
    print(f"MAE(train): {mae_train:.2f}, MAE(test): {mae_test:.2f}")

    return y_train, y_train_pred, y_test, y_test_pred
```

```python
X_slice_cv = X.loc[:, X.std() / X.mean() > 2]
X_slice_std = X.loc[:, X.std() > 5]

fig = plt.figure(figsize=(9, 3))

features = {"raw": X, "std": X_slice_std, "cv": X_slice_cv}

for i, (name, X_) in enumerate(features.items()):
    ax = fig.add_subplot(1, 3, i + 1)
    y_train, y_train_pred, y_test, y_test_pred = train_and_eval(X_slice_cv, y)
    ax.scatter(y_test, y_test_pred, label=name, alpha=0.5, s=10)
    ax.legend()
    ax.plot([-10, 5], [-10, 5], linewidth=1, color="tab:gray")
    ax.set_xlabel("logS(true)")
    ax.set_ylabel("logS(pred)")
    ax.set_xlim(y.min(), y.max())
    ax.set_ylim(y.min(), y.max())
    ax.set_aspect("equal")
    ax.set_xlim(-10, 5)
    ax.set_ylim(-10, 5)
    ax.legend()
    ax.grid()
    ax.label_outer()

plt.tight_layout()
```
