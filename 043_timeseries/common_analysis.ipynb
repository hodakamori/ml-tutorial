```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv(
    "https://raw.githubusercontent.com/aweglteo/tokyo_weather_data/main/data.csv",
    parse_dates=True,
    index_col=0,
)
```

```python
result = seasonal_decompose(
    df["rain"], model="additive", period=7, extrapolate_trend="freq"
)

fig, axes = plt.subplots(4, 1, figsize=(12, 8))
for attr, ax in zip(["observed", "trend", "seasonal", "resid"], axes):
    ax.plot(getattr(result, attr))
    ax.set_title(attr)
plt.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
plot_acf(df["rain"], lags=30, alpha=0.05, ax=ax)
ax.set_xlabel("Lag")
ax.set_ylabel("ACF")
plt.tight_layout()
```
