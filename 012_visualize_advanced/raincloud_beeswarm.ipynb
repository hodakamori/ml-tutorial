```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("../data/delaney-mordred.csv")
sns.set()
```

```python
df
```

```python
import matplotlib.pyplot as plt
import ptitprince as pt
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
pt.RainCloud(data=df, y="logS", orient="h", ax=ax)
```

```python
df["rank"] = pd.cut(df["logS"], 5, labels=["xs", "s", "m", "l", "xl"])
```

```python
import matplotlib.pyplot as plt
import ptitprince as pt
fig = plt.figure(figsize=(9, 9))
for i in range(0, 9):
    try:
        ax = fig.add_subplot(3, 3, i+1)
        pt.RainCloud(data=df, x="rank", y=f"mordred_{i}", orient="h", ax=ax)
    except AttributeError:
        print(f"morderd_{i} was not properly plotted.")
plt.tight_layout()
        
```
