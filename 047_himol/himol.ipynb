```python
import requests

url = "https://deepchemdata.s3-us-west-1.amazonaws.com/datasets/delaney-processed.csv"
save_path = "/content/HiMol/finetune/dataset/esol/raw/esol.csv"

r = requests.get(url)
with open(save_path, "wb") as f:
    f.write(r.content)
```

```python
!cd /content/HiMol/finetune && mkdir -p model_checkpoints && python finetune.py
```

```python
predict_script = """
from rdkit import Chem
from loader import mol_to_graph_data_obj_simple
from model import GNN_graphpred
import torch

# モデル構築
model = GNN_graphpred(
    num_layer=5,
    emb_dim=512,
    num_tasks=1,
    JK="last",
    drop_ratio=0.5,
    gnn_type="gin"
)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.load_state_dict(torch.load("/content/HiMol/finetune/model_checkpoints/esol.pth", map_location=device))
model.to(device)
model.eval()

# 予測対象SMILES
smiles = "CCO"
mol = Chem.MolFromSmiles(smiles)
data = mol_to_graph_data_obj_simple(mol)

# GPUに移動
data.x = data.x.to(device)
data.edge_index = data.edge_index.to(device)
data.edge_attr = data.edge_attr.to(device)
data.batch = torch.zeros(data.x.size(0), dtype=torch.long).to(device)

# 予測
with torch.no_grad():
    pred = model(data.x, data.edge_index, data.edge_attr, data.batch)

print("Prediction:", pred.item() if pred.numel() == 1 else pred.cpu().numpy())
"""

# ファイルに保存
with open("/content/HiMol/finetune/predict.py", "w") as f:
    f.write(predict_script)
```
