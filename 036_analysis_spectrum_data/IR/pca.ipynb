```python
import pandas as pd
```

```python
names = []
for i in range(10):
    names.append(f"Wn_{i}")
    names.append(f"Int_{i}")
df = pd.read_csv("ir_demo.csv", names=names, skiprows=1)
```

```python
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

for i in range(10):
    ax.plot(df[f"Wn_{i}"], df[f"Int_{i}"], linewidth=0.5)
ax.set_xlabel("Wavenumber (cm$^{-1}$)")
ax.set_ylabel("Intensity (a.u.)")
```

```python
import glob

spectra_csv = glob.glob("*.csv")
spectra = []
for csv in spectra_csv:
    df = pd.read_csv(csv, names=names, skiprows=1)
    label = csv[:4]
    for i in range(10):
        data = df[f"Int_{i}"].values.tolist()
        data.insert(0, label)
        spectra.append(data)

spectra = pd.DataFrame(spectra)
spectra.columns = ["label"] + [f"Int_{i}" for i in range(spectra.shape[1] - 1)]
```

```python
spectra
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

spectra_data = spectra.loc[:, spectra.columns != "label"]
sc = StandardScaler()
spectra_data_scaled = sc.fit_transform(spectra_data)

pca = PCA(n_components=5)
pca_components = pca.fit_transform(spectra_data_scaled)
```

```python
unique_labels = list(set(spectra["label"]))
colors = plt.cm.get_cmap("tab10", len(unique_labels))
label_to_color = {label: colors(i) for i, label in enumerate(unique_labels)}

fig, ax = plt.subplots(figsize=(8, 8))
shown_legend = set()
for i, spectrum in enumerate(spectra["label"]):
    color = label_to_color[spectrum]
    if spectrum not in shown_legend:
        ax.scatter(
            pca_components[i, 0], pca_components[i, 1], color=color, label=spectrum
        )
        shown_legend.add(spectrum)
    else:
        ax.scatter(pca_components[i, 0], pca_components[i, 1], color=color)
ax.legend(ncol=2)
ax.set_xlabel("PC 1")
ax.set_ylabel("PC 2")
```

```python
import umap
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

spectra_data = spectra.loc[:, spectra.columns != "label"]
sc = StandardScaler()
spectra_data_scaled = sc.fit_transform(spectra_data)

umap_model = umap.UMAP(n_components=2, random_state=42)
umap_components = umap_model.fit_transform(spectra_data_scaled)
```

```python
unique_labels = list(set(spectra["label"]))
colors = plt.cm.get_cmap("tab10", len(unique_labels))
label_to_color = {label: colors(i) for i, label in enumerate(unique_labels)}

fig, ax = plt.subplots(figsize=(8, 8))
shown_legend = set()
for i, spectrum in enumerate(spectra["label"]):
    color = label_to_color[spectrum]
    if spectrum not in shown_legend:
        ax.scatter(
            umap_components[i, 0], umap_components[i, 1], color=color, label=spectrum
        )
        shown_legend.add(spectrum)
    else:
        ax.scatter(umap_components[i, 0], umap_components[i, 1], color=color)
ax.legend(ncol=3)
plt.show()
```
