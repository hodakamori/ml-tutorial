```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def random_sin(x, a, b, c):
    y = a * np.sin(x - b) + c
    return y
```

```python
a = np.random.normal(-1, 1)
b = np.random.normal(2, 3)
c = np.random.normal(-1, 1)

x = np.linspace(-np.pi, np.pi, 100)
y = random_sin(x, a, b, c)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
y_sum = np.zeros(100)
y_sqsum = np.zeros(100)
for _ in range(10):
    a = np.random.normal(-1, 1)
    b = np.random.normal(2, 3)
    c = np.random.normal(-1, 1)
    x = np.linspace(-np.pi, np.pi, 100)
    y = random_sin(x, a, b, c)
    y_sum += y
    y_sqsum += y**2
    ax.plot(x, y, linewidth=1)
y_mean = y_sum / 10
y_std = y_sqsum / 100 - y_sum / 10**2
ax.plot(x, y_mean, marker="o", linewidth=2, ms=2)
ax.fill_between(x, y1=y_mean + 3 * y_std, y2=y_mean - 3 * y_std, alpha=0.3)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_ylim(-4, 4)
```
