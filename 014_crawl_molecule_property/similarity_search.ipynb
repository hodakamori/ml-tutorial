```python
import pubchempy as pcp
```

```python
# CIDから化合物情報を取得
compound = pcp.Compound.from_cid(2244)
print(compound.molecular_formula)
```

```python
# 名前から化合物を検索
compounds = pcp.get_compounds("caffeine", "name")
for compound in compounds:
    print(compound.molecular_weight)
```

```python
query_smiles = "CC(=O)OC1=CC=CC=C1C(=O)O"
similar = pcp.get_compounds(
    query_smiles, "smiles", searchtype="similarity", listkey_count=3
)
for compound in similar:
    print(compound.cid)
```

```python
smiles = [compound.canonical_smiles for compound in similar]
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

mol = Chem.MolFromSmiles(query_smiles)
Draw.MolToImage(mol)
```

```python
mols = [Chem.MolFromSmiles(smi) for smi in smiles]
Draw.MolsToGridImage(mols)
```
