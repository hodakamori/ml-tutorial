```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
df = pd.read_csv("../data/delaney-mordred.csv")
X = df.loc[:, df.columns.str.startswith("mordred")]
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
X_std
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_std)
feature = pca.transform(X_std)
pd.DataFrame(feature, columns=[f"PCA{i}" for i in range(feature.shape[1])])
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(121)
m = ax.scatter(feature[:, 0], df['logS'].values, alpha=0.5, s=5)
ax.set_xlabel('PCA0')
ax.set_ylabel('logS')
ax = fig.add_subplot(122)
m = ax.scatter(feature[:, 1], df['logS'].values, alpha=0.5, s=5)
ax.set_xlabel('PCA1')
ax.set_ylabel('logS')
ax.label_outer()
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(121)
m = ax.scatter(feature[:, 0], feature[:, 1], alpha=0.5, s=10, c=df["logS"].values, cmap="RdYlBu_r")
fig.colorbar(m, ax=ax, label='logS')
ax.set_xlabel('PCA0')
ax.set_ylabel('PCA1')
plt.tight_layout()
```
