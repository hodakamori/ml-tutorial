```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
df = pd.read_csv("../data/delaney-mordred.csv")
X = df.loc[:, df.columns.str.startswith("mordred")]
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
X_std
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_std)
feature = pca.transform(X_std)
pd.DataFrame(feature, columns=[f"PCA{i}" for i in range(feature.shape[1])])
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(121)
m = ax.scatter(feature[:, 0], df['logS'].values, alpha=0.5, s=5)
ax.set_xlabel('PCA0')
ax.set_ylabel('logS')
ax = fig.add_subplot(122)
m = ax.scatter(feature[:, 1], df['logS'].values, alpha=0.5, s=5)
ax.set_xlabel('PCA1')
ax.set_ylabel('logS')
ax.label_outer()
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(121)
m = ax.scatter(feature[:, 0], feature[:, 1], alpha=0.5, s=10, c=df["logS"].values, cmap="RdYlBu_r")
fig.colorbar(m, ax=ax, label='logS')
ax.set_xlabel('PCA0')
ax.set_ylabel('PCA1')
ax.set_title("Linear PCA")
plt.tight_layout()
```

```python
from sklearn.manifold import TSNE
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(121)
tsne_model = TSNE(n_components=2)
tsne = tsne_model.fit_transform(X_std)
m = ax.scatter(tsne[:,0], tsne[:,1], alpha=0.5, s=10, c=df["logS"].values, cmap="RdYlBu_r")
fig.colorbar(m, ax=ax, label='logS')
ax.set_xlabel('t-SNE 0')
ax.set_ylabel('t-SNE 1')
ax.set_title("t-SNE")
plt.tight_layout()
```

```python
import umap
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(121)
umap_embedding = umap.UMAP().fit_transform(X_std)
m = ax.scatter(umap_embedding[:,0], umap_embedding[:,1], alpha=0.5, s=10, c=df["logS"].values, cmap="RdYlBu_r")
fig.colorbar(m, ax=ax, label='logS')
ax.set_xlabel('UMAP 0')
ax.set_ylabel('UMAP 1')
ax.set_title("UMAP")
plt.tight_layout()
```
